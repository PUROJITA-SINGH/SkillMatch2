```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Loan prediction - Loan prediction.csv')
```

```python
df.info()
```

Univariate Analysis for Numerical Variables

```python
#Histgram => Loan Amount Term
plt.hist(df['Loan_Amount_Term'])
plt.xlabel('Loan Amount Term')
plt.ylabel('Count')
plt.show()
```

```python
#BoxPlot => Applicant Income
plt.boxplot(df['ApplicantIncome'])
plt.xlabel('Applicant Income')
plt.show()
```

```python
#ViolinPLot => Co Applicant Income
plt.violinplot(df['CoapplicantIncome'])
plt.xlabel('Co-Applicant Income')
plt.ylabel('Count')
plt.show()
```

u for Categorical Var

```python
# Bar Graph => Gender
gender_counts = df['Gender'].value_counts()
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Pie Chart => Dependents
dependents = df['Dependents'].value_counts()
plt.pie(dependents.values, labels=dependents.index, autopct='%1.1f%%')
plt.title('Dependents Distribution')
plt.show()
```

Bivariate A for numerical var

```python
#ScatterPlot => Applicant Income & Loan Amount
plt.scatter(df['ApplicantIncome'], df['LoanAmount'])
plt.title('Scatter Plot of Applicant Income vs Loan Amount')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

```python
#JoinPlot => ApplicantIncome & Loan Amount
sns.jointplot(df, x='LoanAmount', y='ApplicantIncome', kind='scatter')
plt.suptitle('Scatter Plot of Applicant Income vs Loan Amount')
plt.show()
```

bivar for cat var

```python
#Count Plot => Education & Loan Status
sns.countplot(df, x='Education', hue='Loan_Status')
plt.suptitle('Count Plot of Education vs Loan Status')
plt.show()
```

```python
#Bar Plot => Gender & Applicant Income
sns.barplot(df, x='Gender', y='ApplicantIncome')
plt.suptitle('Bar Plot of Gender vs Applicant Income')
plt.show()
```

```python
#Box Plot => Self Employed & Application Income
sns.boxplot(df, x='Self_Employed', y='ApplicantIncome')
plt.suptitle('Box Plot of Self Employed vs Applicant Income')
plt.show()
```

```python
#Violin Plot => Property Area & Loan Amount
sns.violinplot(df, x='Property_Area', y='LoanAmount')
plt.suptitle('Violin Plot of Property Area vs Loan Amount')
plt.show()
```

```python
#Displot => Applicant Income & Married
sns.displot(df, x='ApplicantIncome', hue='Married',kind='kde')
plt.suptitle('     Distribution Plot of Applicant Income vs Married')
plt.show()
```

multivariate analysis

```python
num_df=df.select_dtypes(include=['number'])
sns.heatmap(num_df.corr(),annot=True)
plt.title('Correlation Heatmap for all Numerical Variables')
plt.show()
```
